```python
weight, height = float(input()), float(input())
imt_value = weight / (height**2)
res = str()
if imt_value < 18.5:
    res = 'Недостаточная масса'
elif 18.5 <= imt_value <= 25:
    res = 'Оптимальная масса'
else:
    res = 'Избыточная масса'
print(res)
```

```python
input_num = input()
if len(input_num) == 5:
    print(int(input_num[::-1]))
else:
    print(input_num[0] + input_num[:-6:-1])
```

```python
input_num = input()
rev_input_num = input_num[::-1]
new_str = str()
for i in range(len(rev_input_num)):
    if i != 0 and i % 3 == 0:
        new_str += ','
    new_str += rev_input_num[i]
print(new_str[::-1])             #мой способ
print(f'{int(input_num):,}')     #через f-строку
```

```python
num_cnt = int(input())
first_quarter = 0
second_quarter = 0
third_quarter = 0
fourth_quarter = 0
for _ in range(num_cnt):
    input_list = input().split()
    if int(input_list[0]) > 0 and int(input_list[1]) > 0:
            first_quarter += 1
    elif int(input_list[0]) < 0 and int(input_list[1]) > 0:
            second_quarter += 1
    elif int(input_list[0]) < 0 and int(input_list[1]) < 0:
            third_quarter += 1
    elif int(input_list[0]) > 0 and int(input_list[1]) < 0:
            fourth_quarter += 1
print(f'Первая четверть: {first_quarter}')
print(f'Вторая четверть: {second_quarter}')
print(f'Третья четверть: {third_quarter}')
print(f'Четвертая четверть: {fourth_quarter}')
```
