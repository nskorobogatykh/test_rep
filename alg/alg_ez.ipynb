```python
def get_sum(a, b) -> int:
    return a + b

print(get_sum(5, 7))
```

```python
def mid_num(a, b, c):
    lst = [a,b,c]
    min_lst = min(lst)
    max_lst = max(lst)
    lst.remove(min_lst)
    lst.remove(max_lst)
    return lst[0]

print(mid_num(1, 2, 3))
```
